```python
import matplotlib.pyplot as plt
import pandas as pd

df_b = pd.read_csv('assets/bike.csv')

hourly_avg = df_b.groupby('hr')[['casual', 'registered']].mean()
hourly_avg_normalized = hourly_avg / hourly_avg.max().max()

#  
plt.figure(figsize=(10, 6))
plt.plot(hourly_avg_normalized.index, hourly_avg_normalized['casual'], '--o', color='blue', label='Casual')
plt.plot(hourly_avg_normalized.index, hourly_avg_normalized['registered'], '-x', color='orange', label='Registered')

#  
plt.title('Average Bike Rentals by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Normalized Average Rentals')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
df = pd.read_csv('assets/bike.csv')

#     
seasonal_totals = df.groupby('season')['cnt'].sum()

#   
seasons = ['Spring', 'Summer', 'Fall', 'Winter']

#  
plt.figure(figsize=(8, 6))
plt.bar(seasons, seasonal_totals, width=0.6, color=['blue', 'green', 'red', 'gray'])

#  
plt.title('Total Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Total Rentals')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
df = pd.read_csv('assets/tips.csv')

#   
lunch_data = df[df['time'] == 'Lunch']
dinner_data = df[df['time'] == 'Dinner']

#  
plt.figure(figsize=(10, 6))
plt.scatter(lunch_data['total_bill'], lunch_data['tip'], color='blue', alpha=0.6, label='Lunch', s=100)
plt.scatter(dinner_data['total_bill'], dinner_data['tip'], color='orange', alpha=0.6, label='Dinner', s=100)

#  
plt.title('Total Bill vs Tip by Time', fontweight='bold')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.grid(linestyle='--', color='gray', alpha=0.7)
plt.show()
```
